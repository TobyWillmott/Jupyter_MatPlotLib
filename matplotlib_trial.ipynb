# Python plotting with matplotlib
- Basic plots


```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
data = [4,5,6,7,8]
ax.bar([1,2,3,4,5], data);
```

```python
test_1 = [5, 3, 7, 9, 8, 3]
test_2 = [5, 8, 8, 2, 7, 6]
x = np.arange(len(test_1))
students = ["Alex", "Katie", "Maya", "Felix", "Lauren", "Arjun"]

fig, axs = plt.subplots(2, 2, figsize=(12,10))

# Bar chart
axs[0,0].bar(students, test_1)
axs[0,0].set_title('Bar Graph')

# Line graph
axs[0,1].plot(x, test_1, label="Test 1", marker='*', linestyle=":")
axs[0,1].plot(x, test_2, label="Test 2", marker='o', linestyle="-.")
axs[0,1].set_title('Line Graph')
axs[0,1].legend()

# Scatter plot
axs[1,0].grid()
axs[1,0].scatter(test_1, test_2, s=60)
axs[1,0].set_title('Scatter Plot')
axs[1,0].set_xlabel('Test 1 score')
axs[1,0].set_ylabel('Test 2 score')

# Stacked bar chart
axs[1,1].barh(students, test_1, label="Test 1", color='indigo')
axs[1,1].barh(students, test_2, label="Test 2", left=test_1, color="violet")
axs[1,1].legend()
axs[1,1].grid(axis='x')
axs[1,1].set_axisbelow(True)
axs[1,1].set_title('Vertical Stacked Bar Chart')
axs[1,1].set_xlabel('Total test scores');
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)


plt.plot(x,y,x,z)
plt.title("Sine and cosine graph")
plt.legend(["sin(x)", "cos(x)"])
plt.show()
```

# stacked vertical line graph

```python
import matplotlib.pyplot as plt
names = ["Alex","Katie", "Maya", "Felix", "Lauren","Arjun"]

sizes = [10,70,20,20,20,90]
explode = (0,0.1,0,0,0,0)
fig,ax = plt.subplots()
ax.pie(sizes, labels=names,explode=explode, autopct="%1.1f%%",colors =["yellow", "blue", "purple","rosybrown", "red", "green"], shadow=True, startangle=0);

```

```python
species = ('Tortoise', 'Lizard', 'Snake')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of reptiles by sex')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(10801)


x = np.arange(0.0, 50.0, 0.4)
y = x ** 1.1 + np.random.rand(*x.shape) * 50.0
sizes = np.random.rand(*x.shape) * 80 + 500

fig, ax = plt.subplots()
ax.scatter(x, y, sizes, c="green", alpha=0.5, marker=r'$\clubsuit$',
           label="Luck")
ax.set_xlabel("Leprechauns")
ax.set_ylabel("Gold")
ax.legend()
plt.show()
```
